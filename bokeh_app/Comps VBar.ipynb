```python
from bokeh.io import show, output_file, output_notebook
from bokeh.models import ColumnDataSource
from bokeh.palettes import Viridis11
from bokeh.plotting import figure
from bokeh.sampledata.autompg import autompg as df
from bokeh.transform import factor_cmap
import numpy as np

import pandas as pd

df = pd.read_csv('data/sales_comps.csv')
df[['Purchase Price','Finished Lot Value']].astype(float)


df.Year = df.Year.astype(str)
group = df.groupby(['Year','Site Condition'])

source = ColumnDataSource(group)

index_cmap = factor_cmap('Year_Site Condition', palette=Viridis11, factors=sorted(df['Site Condition'].unique()), end=1)
cyl_cmap = factor_cmap('Year_Site Condition', palette=Viridis11, factors=sorted(df.Year.unique()))

p = figure(plot_height=350, x_range=group, title="Annual Acquisitions by Site Condition",
           toolbar_location=None, tools="")

p.vbar(x='Year_Site Condition', top='Lot Count_mean', width=.9, source=source,alpha=.8,
       line_color=cyl_cmap, fill_color=index_cmap)

p.y_range.start = 0
p.xgrid.grid_line_color = None
#p.xaxis.axis_label = "some stuff"
p.xaxis.major_label_orientation = 1.2
p.outline_line_color = None

output_notebook()
show(p)
```
