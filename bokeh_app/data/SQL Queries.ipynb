```python
import sqlite3 as sql
import pandas as pd
from os.path import join, dirname
from sqlalchemy.types import *

markets = pd.read_csv('markets.csv')
markets = markets.iloc[:,0:8]
markets.head()
con = sql.connect('Market.db')
datatypes = {
    'Market': 'VARCHAR(55)',
    'City': 'VARCHAR(55)',
    'State': 'CHAR(2)',
    'Zip Code': 'VARCHAR(5)',
    'Market Tier': 'CHAR(1)',
    'Sub-Market': 'VARCHAR(55)',
    'Lat': 'DECIMAL(8,6)',
    'Long': 'DECIMAL(9,6)'
}
markets.to_sql('markets', con, if_exists='replace', dtype=datatypes)
con.close()
```
